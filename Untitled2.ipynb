```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
net = pd.read_csv('netflix_titles.csv')
net.head()
```

```python
movie = net.loc[net.type == 'Movie'].count()[0]
tvshow = net.loc[net.type=='TV Show'].count()[0]
plt.style.use('default')
labels=['Movie','TV Show']
plt.bar(labels,[movie,tvshow])
plt.legend()
plt.show()
```

```python
movie = net[net['type']=='Movie']
movie = movie.dropna(subset=['duration'])
movie['dura']=movie['duration'].str.split(' ').str[0].astype(int)
movie['dura'].mean()
```

```python
tv=net.loc[net['type']=='TV Show']
tv['tv_dur']=tv['duration'].str.split(' ').str[0].astype(int)
tv.groupby('title').agg({'tv_dur':'sum'}).sort_values('tv_dur',ascending=False)
```

```python
# Top 10 most frequent actor collaborations.
from itertools import combinations
df = net.dropna(subset=['cast'])
df['actor']=df['cast'].str.split(', ')
pairs=df['actor'].apply(lambda x:list(combinations(x,2)))
pairs.explode().value_counts().head(10)
```

```python
gener = net.copy()
gener = gener.dropna(subset=['country','listed_in'])
gener['gen']=gener['listed_in'].str.split(', ')
gener=gener.explode('gen')

gener['country'] = gener['country'].str.split(', ')
gener = gener.explode('country')

gener['country'] = gener['country'].str.strip()   # remove spaces
gener = gener[gener['country'] != '']
pop_gen = gener.groupby(['country','gen']).size().reset_index(name='count').sort_values('count',ascending=False)
pop_gen.head()
```
